```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
```

```python
data = load_iris()
X = data.data
y=data.target
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)
```

```python
scatter = plt.scatter(X_pca[:,0],X_pca[:,1],c=y)
```
