```python
from sklearn.linear_model import LogisticRegression, LinearRegression
from sklearn.metrics import mean_squared_error, precision_score, recall_score, accuracy_score, f1_score
import numpy as np
```

```python
X = [
    [1,2,3,4],
    [5,2,4,4],
    [11,22,13,34],
    [51,32,43,42],
    
    [10,20,30,40]
]
y_r = [
    1,2,3,4,5
]
y_c = [
    1,0,1,0,1
]

test = [[15,27,6,40],]

lin_mod = LinearRegression()
log_mod = LogisticRegression()

lin_mod.fit(X,y_r)
log_mod.fit(X,y_c)

y_r_pred = lin_mod.predict(X)
y_c_pred = log_mod.predict(X)

print("Accuracy: ", accuracy_score(y_c,y_c_pred))
print("Precision: ", precision_score(y_c,y_c_pred))
print("Recall: ", recall_score(y_c,y_c_pred))
print("F1: ", f1_score(y_c,y_c_pred))

print("MSE: ", mean_squared_error(y_r,y_r_pred))
print("RMSE: ", np.sqrt(mean_squared_error(y_r,y_r_pred)))

print("Output for test:")
print("Linear:")
print(lin_mod.predict(test))
print("Logistic:")
print(log_mod.predict(test))
```
