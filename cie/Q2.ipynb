```python
# Find s
def find_s(training_data):
    hypothesis = [None]*(len(training_data[0])-1)
    for ex in training_data:
        if(ex[-1]=='No'):
            continue
        for i in range(len(hypothesis)):
            if hypothesis[i]==None:
                hypothesis[i] = ex[i]
            elif hypothesis[i]!=ex[i]:
                hypothesis[i] = '?'
    return hypothesis
```

```python
training_data = [
    ['Sunny','Warm','Normal', 'Strong', 'Yes'],
    ['Sunny','Warm','High', 'Strong', 'Yes'],
    ['Sunny','Cold','High', 'Strong', 'No'],
    ['Sunny','Warm','Normal', 'Weak', 'Yes'],
]

print(find_s(training_data))
```
