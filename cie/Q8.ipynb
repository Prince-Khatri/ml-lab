```python
from sklearn.cluster import AgglomerativeClustering, BisectingKMeans
from scipy.cluster.hierarchy import linkage, dendrogram
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt

# Dataset
X, y = make_blobs(
    n_samples=100,
    centers=3,
    cluster_std=1.0,
    random_state=42
)

# AGNES (Bottom-Up)
agnes = AgglomerativeClustering(n_clusters=3)

# DIANA Approximation (Top-Down)
diana = BisectingKMeans(n_clusters=3, random_state=42)

# Fit and Predict
a_l = agnes.fit_predict(X)
d_l = diana.fit_predict(X)

# -----------------------------
# AGNES Clustering Output
# -----------------------------
plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=a_l)
plt.title("AGNES Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# -----------------------------
# DIANA Clustering Output
# -----------------------------
plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], c=d_l)
plt.title("DIANA Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()

# -----------------------------
# Dendrogram for AGNES
# -----------------------------
plt.figure(figsize=(10,5))

Z = linkage(X, method="ward")

dendrogram(Z)

plt.title("Dendrogram (Ward Linkage)")
plt.xlabel("Data Points")
plt.ylabel("Euclidean Distance")
plt.show()
```
